```python
import pandas as pd
```

```python
hitting_df = pd.read_csv('resources/hitting_data.csv')
pitching_df = pd.read_csv('resources/pitching_data.csv')
```

```python
pd.set_option('display.max_columns', None)
hitting_df = hitting_df.add_suffix('-hit')
hitting_df.head()
```

```python
hitting_df = hitting_df.rename(columns={'Year-hit':'Year', 'Tm-hit':'Tm'})
hitting_df.head()
```

```python
pd.set_option('display.max_columns', None)
pitching_df = pitching_df.add_suffix('-pitch')
pitching_df = pitching_df.rename(columns={'Year-pitch':'Year', 'Tm-pitch':'Tm'})
pitching_df.head()
```

```python
hitting_df['Year-Tm'] = hitting_df['Year'].astype(str) + hitting_df['Tm']
hitting_df.head()
```

```python
pitching_df['Year-Tm'] = pitching_df['Year'].astype(str) + pitching_df['Tm']
pitching_df.head()
```

```python
combined_stats_df = hitting_df.merge(pitching_df, on='Year-Tm')
```

```python
pd.set_option('display.max_columns', None)
combined_stats_df.head()
```

```python
combined_stats_df.to_csv('resources/combined_stats.csv')
```
